# Liverpool Champions League Goals
### Importing packages and loading data

```python
# Importing packages
from matplotlib import pyplot as plt
import pandas as pd
import sys, os
sys.path.insert(0, os.path.abspath('../scripts/'))
import footyviz

# Loading in data of Liverpool goals scored
data = pd.read_csv('../datasets/positional_data/liverpool_2019.csv', index_col=('play', 'frame'))
data.tail()
```

We observe every goal included in the dataset

```python
data.index.get_level_values('play').unique()
```

## Bayern 0 - [1] Liverpool
When Liverpool and Bayern Munich met at the Allianz Arena in the 2nd leg of the Round of 16,
a direct ball over the top from Van Dijk to Mane completely opened up the Bayern defence. Mane still had a lot to do,
but a great first touch and then a turn to wrong foot Neuer allowed Mane to direct a delightful chip into the net.
We use voronoi diagrams to map the spaces covered by each team.

```python
# Liverpool 1st goal vs Bayern
LivBay = 'Bayern 0 - [1] Liverpool'
LivBaydf = data.loc[LivBay]
```

With Van Dijk on the ball, we see the space behind the Bayern defence for Mane to exploit by cutting in from the left wing.

```python
fig, ax, dfFrame = footyviz.draw_frame(LivBaydf, t=0)
fig, ax, dfFrame = footyviz.add_voronoi_to_fig(fig, ax, dfFrame)
```

We now observe the moment Mane receives the ball, and the quick recovery of the Bayern defenders a moment later.
Mane receives the ball:

```python
fig, ax, dfFrame = footyviz.draw_frame(LivBaydf, t=4)
fig, ax, dfFrame = footyviz.add_voronoi_to_fig(fig, ax, dfFrame)
```

Mane turns to wrong foot Neuer and prepare a shot

```python
fig, ax, dfFrame = footyviz.draw_frame(LivBaydf, t=6)
fig, ax, dfFrame = footyviz.add_voronoi_to_fig(fig, ax, dfFrame)
```

The ball is in the back of the net

```python
fig, ax, dfFrame = footyviz.draw_frame(LivBaydf, t=8)
fig, ax, dfFrame = footyviz.add_voronoi_to_fig(fig, ax, dfFrame)
```

## Porto 0 - [2] Liverpool
Liverpool faced Porto in the Quarter Finals, beating them 6-1 on aggregate and scoring this goal in the first leg at Anfield:

```python
# Liverpool 2nd goal vs Porto
LivPor = 'Porto 0 - [2] Liverpool'
LivPordf = data.loc[LivPor]
```

The goal starts with a counter-attack with Trent Alexander-Arnold on the ball, who sees Mo Salah on the left wing and plenty of space for him to run into.

```python
fig, ax, dfFrame = footyviz.draw_frame(LivPordf, t=8)
fig, ax, dfFrame = footyviz.add_voronoi_to_fig(fig, ax, dfFrame)
```

By the time Mo Salah reaches the ball, he's straight on goal.

```python
fig, ax, dfFrame = footyviz.draw_frame(LivPordf, t=10.5)
fig, ax, dfFrame = footyviz.add_voronoi_to_fig(fig, ax, dfFrame)
```

This is an interesting point in the play. Mo Salah has his sights on goal, however the simple tap in is lined up for Mane if Salah decided to square it to him.

```python
fig, ax, dfFrame = footyviz.draw_frame(LivPordf, t=11.5)
fig, ax, dfFrame = footyviz.add_voronoi_to_fig(fig, ax, dfFrame)
```

Mo Salah only has eyes for the goal, however, and slots it cooly past the keeper.

```python
fig, ax, dfFrame = footyviz.draw_frame(LivPordf, t=12)
fig, ax, dfFrame = footyviz.add_voronoi_to_fig(fig, ax, dfFrame)
fig, ax, dfFrame = footyviz.draw_frame(LivPordf, t=12.5)
fig, ax, dfFrame = footyviz.add_voronoi_to_fig(fig, ax, dfFrame)
```
